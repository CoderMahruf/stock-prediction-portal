```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Type of plot

## Line Plot

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]
```

```python
plt.plot(x,y);
plt.title("Simple line plot");
plt.xlabel("X-axis table");
plt.ylabel("Y-axis table");
plt.grid(True)
```

## Bar chart

```python
months = ['Jan','Feb','Mar','April']
revenue = [100,200,300,400]
```

```python
plt.figure(figsize=(10,6))
plt.bar(months,revenue);
plt.title("Monthly Revenue Bar Chart");
plt.xlabel("Months");
plt.ylabel("Revenue");
```

## Scatter plot

```python
heights = [65,70,68,72,66]
weights = [150,160,155,180,145]
```

```python
plt.scatter(heights,weights);
plt.title("Heights vs Weights Scatter plot");
plt.xlabel("Heights (inches)");
plt.ylabel("Weights (pounds)");
```

## Subplots and legends

```python
days = [1,2,3,4,5]
original_price = [10,20,25,30,40]
predicted_price = [8,21,25,28,38]
plt.plot(days,original_price, 'b-o', label="Original Price")
plt.plot(days,predicted_price,'r-o',label="Predicted Price")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
```

## Sublots

```python
# First subplots - Original price
plt.subplot(3,1,1)
plt.plot(days,original_price, 'b-o', label="Original Price")
plt.title("Original prices over time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

# Second subplots - Predicted price
plt.subplot(3,1,2)
plt.plot(days,predicted_price,'r-o',label="Predicted Price")
plt.title("Predicted prices over time")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.legend()

# Third subplots - Difference between original & Predicted prices
price_difference = [o - p for o,p in zip(original_price,predicted_price)]
plt.subplot(3,1,3)
plt.bar(days,price_difference,color='purple',label="Price Difference")
plt.title("Difference between original & Predicted prices")
plt.xlabel("Days")
plt.ylabel("Price Difference")
plt.legend()

plt.tight_layout()
```

## Saving the figure 

```python
plt.plot(days,original_price, 'b-o', label="Original Price")
plt.plot(days,predicted_price,'r-o',label="Predicted Price")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("Weekdays")
plt.ylabel("Price")
plt.savefig('example_prediction.png', dpi = 300) # save the image 
```
